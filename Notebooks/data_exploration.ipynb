```python
# global imports
import numpy as np
import pandas as pd

# for graphics
import plotly.graph_objects as go
import plotly.express as px
```

```python
# stocks name
STOCK_NAMES = [f"stock_{i}" for i in range(4)]

# load data
stocks_df = pd.read_csv(
    # path to data
    "Data/train.csv",
    # use 1st col as index
    index_col=0,
    # name of cols
    header=None,
    names=STOCK_NAMES
)
```

```python
stocks_df.info()
```

```python
# plots of stocks
fig = go.Figure(
    data=[
        go.Scatter(
            y=stocks_df[name],
            mode="lines+markers",
            name=name
        )
    for name in STOCK_NAMES]
)

fig.update_layout(
    title="Plots of the stocks indexes",
    xaxis={
        "title": "time"
    },
    yaxis={
        "range": [-0.01, 0.15]
    }
)

fig.show()
```

```python
# distribution of stocks
import plotly.figure_factory as ff


# Group data together
hist_data = [stocks_df[name] for name in STOCK_NAMES]
group_labels = STOCK_NAMES

# Create distplot
fig = ff.create_distplot(hist_data, group_labels)

fig.update_layout(
    title="Distribution of stocks",
    xaxis={
        "title": "values"
    },
    yaxis={
        "title": "count",
        "range": [-2, 80]
    }
)
fig.show()
```
